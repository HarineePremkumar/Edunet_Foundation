```python
!apt-get install -y xvfb # Install Xvfb
!pip install pyvirtualdisplay # Install pyvirtualdisplay
!pip install diffusers transformers accelerate scipy safetensors customtkinter Pillow

import tkinter as tk
import customtkinter as ctk
from PIL import Image, ImageTk
import torch
from diffusers import StableDiffusionPipeline
import io
from google.colab import files
from pyvirtualdisplay import Display # Import Display

def run_stable_diffusion_gui():
    # Start virtual display
    display = Display(visible=0, size=(532, 632))
    display.start()

    app = tk.Tk()
    app.geometry("532x632")
    app.title("Stable Bud")
    ctk.set_appearance_mode("dark")

    prompt = ctk.CTkEntry(master=app, height=40, width=512,
                          font=("Arial", 20), text_color="black", fg_color="white")
    prompt.place(x=10, y=10)

    lmain = ctk.CTkLabel(master=app, height=512, width=512)
    lmain.place(x=10, y=110)

    device = "cuda" if torch.cuda.is_available() else "cpu"
    torch_dtype = torch.float16 if torch.cuda.is_available() else torch.float32

    pipe = StableDiffusionPipeline.from_pretrained(
        "runwayml/stable-diffusion-v1-5", torch_dtype=torch_dtype # using a more robust model
    ).to(device)

    def generate():
        prompt_text = prompt.get()
        image = pipe(prompt_text, guidance_scale=8.5).images[0]

        # Convert PIL Image to PhotoImage
        img_byte_arr = io.BytesIO()
        image.save(img_byte_arr, format='PNG')
        img_byte_arr = img_byte_arr.getvalue()
        img = ImageTk.PhotoImage(data=img_byte_arr)

        lmain.configure(image=img)
        lmain.image = img  # Keep a reference to avoid garbage collection

        # Save and offer download
        image.save('generatedimage.png')
        files.download('generatedimage.png')

    trigger = ctk.CTkButton(master=app, height=40, width=120,
                            font=("Arial", 20), text_color="white", fg_color="blue",
                            text="Generate", command=generate)
    trigger.place(x=206, y=60)

    app.mainloop()

    # Stop virtual display when done
    display.stop()

run_stable_diffusion_gui()
```
